```python
for i in range(len(X_resname)):
    each = X_resname[i]
    new_string = each[3:]
    actual_res_x.append(int(new_string))
    final_res_x.append(each[:3])


for i in range(len(actual_res_x)):
    if (actual_res_x[i] >= 1 and actual_res_x[i] < 295):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 0))
        Final_Residue_X.append(tttx)
    elif (actual_res_x[i] >= 295 and actual_res_x[i] < 430):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 294))
        Final_Residue_X.append(tttx)
    elif (actual_res_x[i] >= 430 and actual_res_x[i] < 532):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 429))
        Final_Residue_X.append(tttx)
    elif (actual_res_x[i] >= 532 and actual_res_x[i] < 660):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 531))
        Final_Residue_X.append(tttx)
    elif (actual_res_x[i] >= 660 and actual_res_x[i] < 782):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 656))
        Final_Residue_X.append(tttx)
    elif (actual_res_x[i] >= 782 and actual_res_x[i] < 917):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 781))
        Final_Residue_X.append(tttx)
    elif (actual_res_x[i] >= 917 and actual_res_x[i] < 1019):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 916))
        Final_Residue_X.append(tttx)
    elif (actual_res_x[i] >= 1019 and actual_res_x[i] < 1147):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 1018))
        Final_Residue_X.append(tttx)
    elif (actual_res_x[i] >= 1147 and actual_res_x[i] < 1269):
        tttx = final_res_x[i] + (str(actual_res_x[i] - 1143))
        Final_Residue_X.append(tttx)

for i in range(len(Final_Residue_X)):
    subunit_residue_X.append(subunits_X[i] + '-' + Final_Residue_X[i])
```
